## 1. What are the key differences between SQL and NoSQL databases?
**Answer:** SQL databases are relational and table-based, while NoSQL databases are non-relational and flexible. SQL uses fixed schemas; NoSQL supports dynamic schemas.

## 2. What makes MongoDB a good choice for modern applications?
**Answer:** MongoDB offers scalability, flexible schema, high performance, JSON-like documents, and cloud support.

## 3. Explain the concept of collections in MongoDB.
**Answer:** Collections are groups of MongoDB documents, similar to tables in SQL but schema-flexible.

## 4. How does MongoDB ensure high availability using replication?
**Answer:** MongoDB uses replica sets with primary and secondary nodes for failover and redundancy.

## 5. What are the main benefits of MongoDB Atlas?
**Answer:** Atlas provides cloud hosting, backups, scaling, monitoring, and security.

```
!pip install pymongo pandas

from pymongo import MongoClient
import pandas as pd

# Replace with your MongoDB Atlas URI
MONGO_URI = "mongodb+srv://username:password@cluster.mongodb.net/"
client = MongoClient(MONGO_URI)

db = client["SuperstoreDB"]
orders = db["Orders"]

df = pd.read_csv('/content/superstore.csv')
orders.delete_many({})
orders.insert_many(df.to_dict('records'))

print("Documents inserted:", orders.count_documents({}))
```

## 1. Retrieve and print all documents from the Orders collection

```
for doc in orders.find().limit(10):
    print(doc)
```

## 2. Count total documents in Orders collection

```
print(orders.count_documents({}))
```

## 3. Fetch all orders from West region

```
for doc in orders.find({'Region':'West'}).limit(10):
    print(doc)
```

## 4. Find orders where Sales > 500

```
for doc in orders.find({'Sales':{'$gt':500}}).limit(10):
    print(doc)
```
